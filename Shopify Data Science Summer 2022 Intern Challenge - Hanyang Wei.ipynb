On Shopify, we have exactly 100 sneaker shops, and each of these shops sells only one model of shoe. We want to do some analysis of the average order value (AOV). When we look at orders data over a 30 day window, we naively calculate an AOV of $3145.13. Given that we know these shops are selling sneakers, a relatively affordable item, something seems wrong with our analysis. 


```python
import pandas as pd

# First we will read in the dataset and take a look at it
```

```python
df = pd.read_csv('ShopifySneakers.csv')
```

```python
df.head(20)
```

```python
# To confirm the average number, we will calculate the average of order_amount

average_order_value = df['order_amount'].mean()
print(round(average_order_value, 2))
```

Analyzing this dataframe, we see that this average order value is exceedingly high. However, there exists a column call total_items, which is a count for the number of items sold per order. This shows us that customers are buying, at times, more than 1 item per order, giving us this inflated value. In fact, we can see that for order_id 15, we have an order amount of $704000! This value is very high but looking at total_items, we see that this customer actually bought 2000 pairs from this store. We can use the sum of the total_items divided by the total order_amount value to calculate average __item__ value instead of average __order__ value.

```python
average_item_value = df['order_amount'].sum()/df['total_items'].sum()
print(round(average_item_value, 2))
```

We get an average item value of $357.92 which, although may be high for some people for a pair of sneakers, is much more reasonable as an answer.

b. What is the last name of the employee with the most orders?

Answer: Handel (Customer ID = 20)

    SELECT *, COUNT(Orders.CustomerID) FROM Orders
    JOIN Customers
    ON Customers.CustomerID = Orders.CustomerID
    GROUP BY Orders.CustomerID
    ORDER BY COUNT(Orders.CustomerID) DESC;

c. What product was ordered the most by customers in Germany?

Answer: Boston Crab Meat (160 units)

    SELECT OrderDetails.ProductID, OrderDetails.OrderID, Orders.OrderID, Customers.CustomerID, Orders.CustomerID, Country, ProductName, SUM(Quantity) FROM OrderDetails, Orders, Customers, Products
    WHERE OrderDetails.OrderID = Orders.OrderID
    AND Customers.CustomerID = Orders.CustomerID
    AND OrderDetails.ProductID = Products.ProductID
    AND Country = 'Germany'
    GROUP BY OrderDetails.ProductID
    ORDER BY SUM(Quantity) DESC;
